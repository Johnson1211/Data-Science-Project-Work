```python
import yfinance as yf
import pandas as pd
```

```python
ticker_symbol = "EURUSD=X"
start_date = "2000-01-01"
end_date = "2025-01-01"
```

```python
data = yf.download(
    ticker_symbol,
    start=start_date,
    end=end_date,
    interval="1d"
)
```

```python
df = pd.DataFrame(data)
```
